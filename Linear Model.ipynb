```python
import pandas as pd
import numpy as np
import sklearn.linear_model as lm
#predict NIFTY value based on low high close and volume
#read data
df = pd.read_csv(r'C:\Datasets\NIFTY22JANFUT.csv')
df.set_index('dateTime', inplace=True)
#add new column with previous day close
df['PrevClose'] = df['close'].shift(1)
#add new column with previous day volume
df['PrevVolume'] = df['volume'].shift(1)

#drop rows with NaN
df.dropna(inplace=True)

#train model
model = lm.LinearRegression()
model.fit(df[['low', 'high', 'PrevClose', 'PrevVolume']], df['close'])
#predict
df['Predicted'] = model.predict(df[['low', 'high', 'PrevClose', 'PrevVolume']])
#calculate error
df['Error'] = df['Predicted'] - df['close']
#calculate mean absolute error
mae = np.mean(np.abs(df['Error']))
print('Mean Absolute Error: ', mae)
#calculate mean squared error
mse = np.mean(np.square(df['Error']))
print('Mean Squared Error: ', mse)
#print coefficients
print('Coefficients: ', model.coef_)
```
